
### Задание 1.4

Напишите функцию, которая будет переворачивать целые числа, то есть записывать их цифры в обратном порядке. Например, число 734 превратится в 437. Покажите работу функции на примере.



```python
# определяем функцию
def revers(n):
    
    # создаем пустую переменную для записи 
    reversed = ''
    
    # циклом пробегаем по переменной n, записываем каждый полученный символ в начале строки в переменную reversed
    for x in n:
        reversed = x + reversed
    
    # возвращаем результат    
    return reversed

# записываем значение в переменную n и reversed_num   
n = input('Пожалуйста, введите значение:')
reversed_num = revers(n)

# возвращаем результат
print(f'Изначальное значение: {n}')
print(f'перевернутое значение: {reversed_num}')
```

```python
# пока ничего кроме регулярок на ум не приходит для решения этой задачи

# испортируем библиотеку для работы с регулярными выражениями
import re

locality_list = ['Санкт-Петербург', 'посёлок Шушары', 'городской посёлок Янино-1',
       'посёлок Парголово', 'посёлок Мурино', 'Ломоносов', 'Сертолово',
       'Петергоф', 'Пушкин', 'деревня Кудрово', 'Коммунар', 'Колпино',
       'поселок городского типа Красный Бор', 'Гатчина', 'поселок Мурино',
       'деревня Фёдоровское', 'Выборг', 'Кронштадт', 'Кировск',
       'деревня Новое Девяткино', 'посёлок Металлострой',
       'посёлок городского типа Лебяжье',
       'посёлок городского типа Сиверский', 'поселок Молодцово',
       'поселок городского типа Кузьмоловский',
       'садовое товарищество Новая Ропша', 'Павловск',
       'деревня Пикколово', 'Всеволожск', 'Волхов', 'Кингисепп',
       'Приозерск', 'Сестрорецк', 'деревня Куттузи', 'посёлок Аннино',
       'поселок городского типа Ефимовский', 'посёлок Плодовое',
       'деревня Заклинье', 'поселок Торковичи', 'поселок Первомайское',
       'Красное Село', 'посёлок Понтонный', 'Сясьстрой', 'деревня Старая',
       'деревня Лесколово', 'посёлок Новый Свет', 'Сланцы',
       'село Путилово', 'Ивангород', 'Мурино', 'Шлиссельбург',
       'Никольское', 'Зеленогорск', 'Сосновый Бор', 'поселок Новый Свет',
       'деревня Оржицы', 'деревня Кальтино', 'Кудрово',
       'поселок Романовка', 'посёлок Бугры', 'поселок Бугры',
       'поселок городского типа Рощино', 'Кириши', 'Луга', 'Волосово',
       'Отрадное', 'село Павлово', 'поселок Оредеж', 'село Копорье',
       'посёлок городского типа Красный Бор', 'посёлок Молодёжное',
       'Тихвин', 'посёлок Победа', 'деревня Нурма']
       
# определяем функцию
def locality_extractor(locality_list):
    # пустой список для записи обработанных навзаний
    extracted_localities = []
    
    # Регулярное выражение
    pattern = r'(?:поселок городского типа|посёлок городского типа|город|городской посёлок|посёлок|поселок|деревня|село|садовое товарищество)\s+(.*)'

    # цикл пробегет переменной по всему списку
    for loc in locality_list:
        match = re.search(pattern, loc) # поиск совпадений
        if match:
            # Добавляем только название населённого пункта
            extracted_localities.append(match.group(1).strip())
        else:
            # Если нет совпадений, добавляем само название
            extracted_localities.append(loc.strip())
    
    return extracted_localities

# Записываем в переменную очищенный список
clean_loc = locality_extractor(locality_list)

# Выводим результат
print(clean_loc)
```
